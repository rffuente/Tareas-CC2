```python
## Generación de Matrices

def nMatrix(s,t):
    np.random.seed(s)
    B=np.random.rand(t,t)
    return B

def vtest(s,t):
    np.random.seed(s)
    V=np.random.rand(t,1)
    return V
## Matriz vector

def mvect(m,v):
    k=v.size
    aux=np.zeros((k,1))
    for i in range(0,k):
        for j in range(0,k):
            aux[i]+=m[i,j]*v[j]
    return aux
            
            
```

```python
#3.1
ne=np.zeros((10000,1))
for i in range (0,10000):
    m=nMatrix(i+1,100)
    v=vtest(i+1,100)
    pp=mvect(m,v)
    ppn=np.dot(m,v)
    kl=pp-ppn
    ne[i]=np.linalg.norm(kl)
    #print(ne[i])
plt.hist(ne, bins='auto')  
plt.title("Histograma norma del error")
plt.show()

```
